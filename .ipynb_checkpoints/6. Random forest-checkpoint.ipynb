```python
# DOWNLOADING DATA

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

tomato_12 = pd.read_csv('Dataframes/tomato_12.csv') # or index_col = 0 instead index = False in 'qjuices'

tomato_12
```

```python
# SETTING DATA TO TARGET

target_sku = '201676'

sales_names_to_shift = [col for col in tomato_12.columns if 'salesvolume' in col and target_sku not in col]

tomato_12.loc[:, sales_names_to_shift] = tomato_12.loc[:, sales_names_to_shift].shift(1)
tomato_12 = tomato_12.iloc[1:, :]

for colname in sales_names_to_shift:
    tomato_12.rename(columns = {colname: colname + '_lag'}, inplace = True)

tomato_12.rename(columns = {'salesvolume_' + target_sku: 'target', 'av_price_' + target_sku: 'price_of_target'}, inplace = True)
tomato_12.reset_index(drop = True, inplace = True)

print('Corr between target and it\'s price: {}'.format(round(tomato_12.corr().loc['target', 'price_of_target'], 4)))

tomato_12
```

```python
# ADDING more LAGS

names_to_shift = [col for col in tomato_12.columns if 'salesvolume' in col and target_sku not in col] + \
                 [col for col in tomato_12.columns if 'price' in col and target_sku not in col] + \
                 ['target']

for i in [1, 2, 6, 7]:
    for col in names_to_shift:
        tomato_12[col + str(i)] = tomato_12[col].shift(i)

tomato_12 = tomato_12[7:]

tomato_12.reset_index(drop = True, inplace = True)

tomato_12
```

```python
# SETTING X, y

date = tomato_12['date']
X = tomato_12.drop(['target', 'date'], axis = 1)
y = tomato_12['target']
y01 = y.apply(lambda x: 0 if x == 0 else 1)

def timesplit(X, y, train_size = 0.8):
    n = len(y)
    n_train = int(n * train_size)
    X_train, X_holdout = X[:n_train], X[n_train:]
    y_train, y_holdout = y[:n_train], y[n_train:]
    return X_train, X_holdout, y_train, y_holdout

X_train, X_test, y_train, y_test = timesplit(X, y)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

A = []
for depth in [*range(2, 5, 1)] + [*range(5, 20, 3)] + [*range(20, 35, 5)]:
    for number in [*range(3, 8, 1)] + [*range(8, 55, 5)]:
        regressor = RandomForestRegressor(n_estimators = number, random_state = 0,
                                          max_depth = depth, criterion = 'mae',
                                          max_features = 'auto', n_jobs = -1)
        regressor.fit(X_train, y_train)
        y_pred_train = regressor.predict(X_train)
        y_pred_test = regressor.predict(X_test)
        A.append({'max_depth': depth,
                  'n_estimators': number,
                  'MAE_train': metrics.mean_absolute_error(y_train, y_pred_train),
                  'MAE_test': metrics.mean_absolute_error(y_test, y_pred_test),
                  'MSE_train': metrics.mean_squared_error(y_train, y_pred_train),
                  'MSE_test': metrics.mean_squared_error(y_test, y_pred_test)}
                 )
        print(depth, number, 'done')

pd.DataFrame(A)
```

```python
import plotly.express as px

px.line(data_frame = pd.DataFrame(A), x = 'n_estimators', y = 'MAE_test', color = 'max_depth').show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

regressor = RandomForestRegressor(n_estimators = 3, random_state = 0,
                                  max_depth = 3, criterion = 'mae',
                                  max_features = 'auto', n_jobs = -1)

regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
```

```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x = X_test.index, y = [*map(round, y_pred)],
                    mode = 'lines+markers',
                    name = 'Predicted counts',
                         marker = dict(color = '#FF3100', size = 5),
                         line = dict(color = '#FF3100', width = 1.5)
                         ))
fig.add_trace(go.Scatter(x = X_test.index, y = y_test,
                    mode = 'lines+markers',
                    name = 'Actual counts',
                         marker = dict(color = '#00B74A', size = 5),
                         line = dict(color = '#00B74A', width = 1.5)))

fig.show()
```

```python
pd.DataFrame({'date': date[y_train.index], 'predicted': [*map(round, regressor.predict(X_train))], 'real': y_train})\
    .to_csv('Results/Decision trees/Random Forest train.csv')

pd.DataFrame({'date': date[y_test.index], 'predicted': [*map(round, y_pred)], 'real': y_test})\
    .to_csv('Results/Decision trees/Random Forest test.csv')
```

```python
# https://habr.com/ru/company/otus/blog/527554/
# https://catboost.ai/en/docs/concepts/python-reference_catboostregressor
from catboost import CatBoostRegressor

GBDTmodel = CatBoostRegressor(
    iterations = 200,
    learning_rate = 0.15,
    loss_function = 'MAE',
    use_best_model = True)

GBDTmodel.fit(X_train, y_train,
          eval_set = (X_test, y_test),
          verbose = 40,
          plot = True
)

print(f"Model is fitted: {str(GBDTmodel.is_fitted())}")
print(f"Model params: {GBDTmodel.get_params()}")
```

```python
from catboost import Pool, cv
params = {"iterations": 500,
          "depth": 2,
          "loss_function": "MAE",
          "verbose": False}
cv_dataset = Pool(data=X_train,
                  label=y_train)
scores = cv(cv_dataset,
            params,
            fold_count=4, 
            plot="True")
```

```python
GBDTmodel = CatBoostRegressor(
    iterations = 200,
    learning_rate = 0.15,
    loss_function = 'MAE',
    use_best_model = True)

grid = {'learning_rate': [0.03, 0.06, 0.1, 0.15, 0.3, 0.5],
        'depth': [3, 4, 6, 8],
        'l2_leaf_reg': [1, 3, 5, 7, 9]}

grid_search_result = GBDTmodel.grid_search(grid, X = X_train, y = y_train, plot = True)
```
