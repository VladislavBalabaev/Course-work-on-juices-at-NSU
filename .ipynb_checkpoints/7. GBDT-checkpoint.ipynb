```python
# DOWNLOADING DATA

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

tomato_discount = pd.read_csv('Dataframes/tomato_discount.csv')
tomato_12 = pd.read_csv('Dataframes/tomato_12.csv') # or index_col = 0 instead of index = False in 'qjuices'

# tomato_12
```

```python
# SETTING DATA TO TARGET

target_sku = '201676'

sales_names_to_shift = [col for col in tomato_12.columns if 'salesvolume_' in col and target_sku not in col]

tomato_12.loc[:, sales_names_to_shift] = tomato_12.loc[:, sales_names_to_shift].shift(1)
tomato_12 = tomato_12.iloc[1:, :]

for colname in sales_names_to_shift:
    tomato_12.rename(columns = {colname: colname + '_lag'}, inplace = True)

tomato_12.rename(columns = {'salesvolume_' + target_sku: 'target', 'av_price_' + target_sku: 'price_of_target'}, inplace = True)
tomato_12.reset_index(drop = True, inplace = True)

print('Corr between target and it\'s price: {}'.format(round(tomato_12.corr().loc['target', 'price_of_target'], 4)))

# tomato_12
```

```python
# ADDING more LAGS

names_to_shift = [col for col in tomato_12.columns if 'salesvolume_' in col] + \
                 [col for col in tomato_12.columns if 'av_price_' in col] + \
                 [col for col in tomato_12.columns if 'is_supplied_' in col] + \
                 [col for col in tomato_12.columns if 'sales_week_' in col] + \
                 ['target']

for i in [1, 2, 6, 7]:
    for col in names_to_shift:
        tomato_12[col + '_' + str(i)] = tomato_12[col].shift(i)

tomato_12 = tomato_12[7:]

tomato_12.reset_index(drop = True, inplace = True)

# tomato_12
```

```python
# SETTING X, y

date = tomato_12['date']
X = tomato_12.drop(['target', 'date'], axis = 1)
y = tomato_12['target']

def timesplit(X, y, train_size = 0.8):
    n = len(y)
    n_train = int(n * train_size)
    X_train, X_holdout = X[:n_train], X[n_train:]
    y_train, y_holdout = y[:n_train], y[n_train:]
    return X_train, X_holdout, y_train, y_holdout

X_train, X_test, y_train, y_test = timesplit(X, y)
```

```python
# CROSS-VALIDATION OF GBDT

from catboost import CatBoostRegressor

GBDTmodel = CatBoostRegressor(iterations = 500, loss_function = 'MAE')

param_grid = {'learning_rate': [0.005, 0.01, 0.03, 0.06, 0.1, 0.15, 0.3, 0.5, 0.75],
              'depth': [*range(2, 11, 1)] + [13, 15] + [None]}  # 'l2_leaf_reg': [1, 3, 5, 7, 9]

grid_search_result = GBDTmodel.grid_search(param_grid, X = X, y = y, plot = True, train_size = 0.8, cv = 3)

print("\n The best estimator across ALL searched params:\n", grid_search_result.best_estimator_)
print("\n The best score across ALL searched params:\n", grid_search_result.best_score_)
print("\n The best parameters across ALL searched params:\n", grid_search_result.best_params_)
```

```python
# ESTIMATING OF GBDT

from catboost import CatBoostRegressor
from sklearn import metrics

GBDTmodel = CatBoostRegressor(
    iterations = 500,
    depth = None,
    learning_rate = 0.1,
    loss_function = 'MAE',
    use_best_model = True)

GBDTmodel.fit(X_train, y_train,
          eval_set = (X_test, y_test),
          verbose = 50,
          plot = True
)

print(f"Model is fitted: {str(GBDTmodel.is_fitted())}")
# print(f"Model params: {GBDTmodel.get_params()}")

GBDT_train_pred = GBDTmodel.predict(X_train)
GBDT_test_pred = GBDTmodel.predict(X_test)

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, GBDT_test_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, GBDT_test_pred))
```

```python
GBDTmodel.plot_tree(tree_idx = 0)
```

```python
# GBDT train

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=tomato_12.loc[X_train.index, 'date'],
                         y=y_train,
                         mode='lines+markers',
                         name='Actual counts',
                         marker=dict(color='#00A383', size=3.5),
                         line=dict(color='#00A383', width=1.5)))

fig.add_trace(go.Scatter(x=tomato_12.loc[X_train.index, 'date'],
                         # y = [*map(round, GBDT_test_pred)],
                         y=GBDT_train_pred,
                         mode='lines',
                         name='Predicted counts',
                         line=dict(color='#F53D65', width=2.5)
                         ))

fig.add_trace(go.Scatter(x=tomato_12.loc[X_train.index, 'date'],
                         y=tomato_discount[tomato_discount['date']
                         .isin(tomato_12.loc[X_train.index, 'date'])]
                         .loc[:, 'is_discount_201676'] * 50,
                         mode='lines',
                         name='Presence of discount',
                         line=dict(color='#B1F100', width=2.5, dash='dot')  # 'dash'
                         ))

fig.show()
```

```python
# GBDT test

import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(x=tomato_12.loc[X_test.index, 'date'],
                         y=y_test,
                         mode='lines+markers',
                         name='Actual counts',
                         marker=dict(color='#00A383', size=3.5),
                         line=dict(color='#00A383', width=1.5)))

fig.add_trace(go.Scatter(x=tomato_12.loc[X_test.index, 'date'],
                         # y = [*map(round, GBDT_test_pred)],
                         y=GBDT_test_pred,
                         mode='lines',
                         name='Predicted counts',
                         line=dict(color='#F53D65', width=2.5)
                         ))

fig.add_trace(go.Scatter(x=tomato_12.loc[X_test.index, 'date'],
                         y=tomato_discount[tomato_discount['date']
                         .isin(tomato_12.loc[X_test.index, 'date'])]
                         .loc[:, 'is_discount_201676'] * 50,
                         mode='lines',
                         name='Presence of discount',
                         line=dict(color='#B1F100', width=2.5, dash='dot')  # 'dash'
                         ))

fig.show()
```

```python
# SAVING RESULTS OF GBDT

pd.DataFrame({'date': date[y_train.index],
              'predicted': GBDT_train_pred,
              # 'predicted': [*map(round, GBDT_train_pred)],
              'real': y_train})\
    .to_csv('Results/Decision trees/GBDT train.csv', index = False)

pd.DataFrame({'date': date[y_test.index],
              'predicted': GBDT_test_pred,
              # 'predicted': [*map(round, GBDT_test_pred)],
              'real': y_test})\
    .to_csv('Results/Decision trees/GBDT test.csv', index = False)
```
